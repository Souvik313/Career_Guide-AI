```python
import matplotlib.pyplot as plt

# Top 15 job positions
top_positions = df["Position"].value_counts().head(15)

# Plotting 
plt.figure(figsize=(10 , 6))
plt.bar(top_positions.index , top_positions.values)
plt.xticks(rotation=45, ha="right")
plt.title("Top 15 job positions")
plt.xlabel("Position")
plt.ylabel("Count")
plt.show()
```

```python
# Experience distribution
exp_counts = df["Exp Years"].value_counts().sort_index()

# Plotting
plt.figure(figsize=(16 , 12))
plt.bar(exp_counts.index.astype(str) , exp_counts.values)
plt.title("Experience based recruitment distribution")
plt.xlabel("Years of experience")
plt.ylabel("Number of job postings")
plt.xticks(rotation = 45)
plt.show()
```

```python
# English level distribution
eng_level_counts = df["English Level"].value_counts()

# Plotting
plt.figure(figsize=(10 , 6))
plt.bar(eng_level_counts.index.astype(str) , eng_level_counts.values)
plt.title("English level distribution in job postings")
plt.xlabel("Level")
plt.ylabel("Number of job postings")
plt.xticks(rotation = 45 , ha="right")
plt.show()
```

```python
# Top 10 company distribution
company_count = df["Company Name"].value_counts().head(10)
company_count

plt.figure(figsize=(10 , 6))
plt.bar(company_count.index.astype(str) , company_count.values)
plt.title("Company vs respective number of job postings distribution")
plt.xlabel("Company")
plt.ylabel("Number of job postings")
plt.xticks(rotation = 45 , ha="right")
plt.show()
```

```python
# Top 10 primary keyword distribution
keyword_count = df["Primary Keyword"].value_counts().head(10)

plt.figure(figsize=(12 , 6))
plt.bar(keyword_count.index.astype(str) , keyword_count.values)
plt.title("Primary keyword in job description distribution")
plt.xlabel("Primary Keyword")
plt.ylabel("Number of job postings")
plt.xticks(rotation = 45 , ha="right")
plt.show()
```

```python
# Plotting description counts
mean_length = df["description_length"].mean()

plt.figure(figsize=(10, 6))

plt.hist(
    df["description_length"],
    bins=20,
    edgecolor="black"
)
plt.axvline(
    mean_length,
    color="red",
    linestyle="--",
    linewidth=2,
    label=f"Mean = {mean_length:.0f}"
)

plt.title("Distribution of Job Description Length")
plt.xlabel("Description Length (Characters)")
plt.ylabel("Number of Job Postings")

plt.tight_layout()
plt.show()
```
